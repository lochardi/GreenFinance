```python
from pathlib import Path
import xarray as xr

folder = Path("data_cmip6/")
files = sorted(folder.glob("*.nc"))

print(f"{len(files)} fichiers trouvés")

# Utiliser concat_dim='time' pour une meilleure combinaison sur la dimension temps
ds = xr.open_mfdataset(files, concat_dim='time', combine='nested')
print(ds)
print("\n" + "="*50)
print("Structure du dataset :")
print("="*50)
df = ds.to_dataframe().reset_index()
print(df.head(15))
```

```python
print(df.head(15))
```

```python
df_1 = df.dropna()
df_1.to_csv("data_cmip6/data1.csv", index=False)
```

```python
df_2 = df_1[(df_1['lat'] == 40.1029793042494) & (df_1['lon'] == -3.75)]
df_2.to_csv("data_cmip6/data2.csv", index=False)
print(df_2.head(15))
```

```python
# moyenne par jour (en groupant sur la colonne time)
df_3 = df_2.groupby('time').mean().reset_index()
df_3.to_csv("data_cmip6/data3.csv", index=False)
print(df_3.head(15))
```
